```python
# Importation des librairies
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as st
import datetime


from matplotlib import gridspec
from datetime import datetime
```

```python
# Déclaration d'une fonction
def analyseDataFrame(df):
    """Fonction qui prend en arguments un dataframe pour analyse."""

    name =[x for x in globals() if globals()[x] is df][0]
    print("La dimension de DataFrame « " + name  +" » : " + str(df.shape) + " (lignes, colonnes)\n")
    display(df.info())
    display(df.head())
    display(df.describe().T)
    display(df.nunique())
```

```python
# Importation du dataFrame 'df_enrichi'
df_enrichi = pd.read_csv("df_enrichi.csv", index_col="date_2", parse_dates= True)

# Vérifier que la première colonne est une idex
df_enrichi.index = pd.to_datetime(df_enrichi.index)

# Trier sur la colonne date
df_enrichi = df_enrichi.sort_index()

# Visualisualisation de df_enrichi
analyseDataFrame(df_enrichi)
```

```python
df_temp = df_enrichi[['client_id', 'session_id', 'age', 'price']].copy()

df_Age_CAPanierClient = df_temp.groupby(
    ["client_id", 'session_id', "age"]
).agg(
    {
        'price': ["sum"]
    }).reset_index()
df_Age_CAPanierClient.columns = ["".join(x) for x in df_Age_CAPanierClient.columns]
df_Age_CAPanierClient.rename(columns = {"pricesum" : 'CA_par_panier'}, inplace = True)
df_Age_CAPanierClient.head()
```

```python
df_temp = df_Age_CAPanierClient[['client_id','age', 'CA_par_panier']].copy()

df_Age_PanierMoyenClient = df_temp.groupby(
    ["client_id", "age"]
).agg(
    {
        'CA_par_panier': ["mean"]
    }).reset_index()
df_Age_PanierMoyenClient.columns = ["".join(x) for x in df_Age_PanierMoyenClient.columns]
df_Age_PanierMoyenClient.rename(columns = {"CA_par_paniermean" : 'panier_moyen_client'}, inplace = True)
df_Age_PanierMoyenClient.head()
```

```python
sns.set_style("ticks") # Parameters style dict, or one of {darkgrid, whitegrid, dark, white, ticks}

sns.set_context("paper") # Parameters context dict, or one of {paper, notebook, talk, poster}

plt.figure(figsize = (12,5))

_ = sns.histplot(df_Age_PanierMoyenClient["age"], color='red', kde=True, stat="density", linewidth=0)

_ = plt.axvline(x=df_Age_PanierMoyenClient["age"].mean(), label='Moyenne', linestyle='-', linewidth=2, color='red')
_ = plt.axvline(x=df_Age_PanierMoyenClient["age"].median(), label='Médiane', linestyle='--', linewidth=2, color='red')
_ = plt.title("Distribution de l'âge des clients", fontsize=18)
_ = plt.ylabel("Individues (clients)", fontsize=12)
_ = plt.xlabel("Âge", fontsize=12)
_ = plt.legend(prop={'size': 12})
_ = plt.show()

plt.figure(figsize = (12,5))
_ = sns.histplot(df_Age_PanierMoyenClient["panier_moyen_client"], color='blue', binwidth=2,  kde=True, stat="density", linewidth=0)

_ = plt.axvline(x=df_Age_PanierMoyenClient["panier_moyen_client"].mean(), label='Moyenne', linestyle='-', linewidth=2, color='blue')
_ = plt.axvline(x=df_Age_PanierMoyenClient["panier_moyen_client"].median(), label='Médiane', linestyle='--', linewidth=2, color='blue')
_ = plt.title("Distribution du panier moyen client", fontsize=18)
_ = plt.ylabel("Individues (clients)", fontsize=12)
_ = plt.xlabel("Panier moyen client", fontsize=12)
_ = plt.legend(prop={'size': 12})
_ = plt.show()

# _ = plt.tight_layout()

_ = plt.show()
```

```python
# test de normalité Smirnov-Kolmogorov (loi normal)
# H0 : l'échantillon suit une loi normal / l'echantillon a une distribution Gaussienne
# H1 : hypothèse alternative

print("Premier test Smirnov-Kolmogorov -----------------------------------------------------------")
df_name = 'Âge'

stat, p_value = st.kstest(df_Age_PanierMoyenClient["age"], "norm")
print('Statistics=%.3f, p=%.3f' % (stat, p_value))

alpha = 0.05

if p_value > alpha:
    print(f"L'échantillon '{df_name}' semble avoir une distribution Gaussienne (ne pas rejeter H0")
else:
    print(f"L'échantillon '{df_name}' ne semble pas avoir une distribution Gaussienne (rejeter H0")
print()
# Interprétation : si la valeur-p est inférieure à une seuil α qu’on s’est donné (en général, 5%),
# alors on rejette H0.

print("Deuxième test Smirnov-Kolmogorov -----------------------------------------------------------")
df_name = 'Panier moyen client'

stat, p_value = st.kstest(df_Age_PanierMoyenClient["panier_moyen_client"], "norm")
print('Statistics=%.3f, p=%.3f' % (stat, p_value))

alpha = 0.05

if p_value > alpha:
    print(f"L'échantillon '{df_name}' semble avoir une distribution Gaussienne (ne pas rejeter H0)")
else:
    print(f"L'échantillon '{df_name}' ne semble pas avoir une distribution Gaussienne (rejeter H0)")
print()
# Interprétation : si la valeur-p est inférieure à une seuil α qu’on s’est donné (en général, 5%),
# alors on rejette H0.
```

# Test de corrélations

### Pearson’s Correlation Coefficient

```python
x = df_Age_PanierMoyenClient["age"]
y = df_Age_PanierMoyenClient["panier_moyen_client"]

x_name = "Âge"
y_name = "Le panier moyen client"

alpha = 0.05

r_pearson, p_value = st.pearsonr(x, y)

# r de Pearson
print('r de Pearson          :', r_pearson)


# indice de covariation
print('indice de covariation :', np.cov(x, y, ddof=0)[1,0])

# Coefficient de détermination / R2 / r-squared
print('r-squared             : ', r_pearson**2, " (le coefficient de détermination)")

print('p-value               : ', p_value)

print()

if p_value > alpha:
    print("Probablement independente :")
    print(f"Probablement la variable « {x_name} » n'est pas liée à la variable « {y_name} »")
else:
    print("Probablement dependente :")
    print(f"Probablement la variable « {x_name} » est liée à la variable « {y_name} »")
print()
```

### Spearman’s Rank Correlation

```python
x = df_Age_PanierMoyenClient["age"]
y = df_Age_PanierMoyenClient["panier_moyen_client"]

x_name = "Âge"
y_name = "Le panier moyen"

alpha = 0.05

stat, p_value = st.spearmanr(x, y)

print('Spearmanr stat : ', stat)

print('p-value        : ', p_value)

print()

if p_value > alpha:
    print("Probablement independente :")
    print(f"Probablement la variable « {x_name} » n'est pas liée à la variable « {y_name} »")
else:
    print("Probablement dependente :")
    print(f"Probablement la variable « {x_name} » est liée à la variable « {y_name} »")
print()
```

```python
x = df_Age_PanierMoyenClient["age"]
y = df_Age_PanierMoyenClient["panier_moyen_client"]

x_name = "Âge"
y_name = "Le panier moyen client"

alpha = 0.05

stat, p_value = st.kendalltau(x, y)

print('Kendalltau stat : ', stat)

print('p-value         : ', p_value)

print()

if p_value > alpha:
    print("Probablement independente :")
    print(f"Probablement la variable « {x_name} » n'est pas liée à la variable « {y_name} »")
else:
    print("Probablement dependente :")
    print(f"Probablement la variable « {x_name} » est liée à la variable « {y_name} »")
print()
```
